```python
# Importing the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading of the file of our datset

data = pd.read_csv('winequality-red.csv')
```

```python
# Visualization of top 10 records

data.head(10)
```

```python
# Classes of target value - quality
set(data.quality)
```

```python
# Description of the given dataset
data.describe()
```

```python
# To check for missing values(Although there are none)
data.isnull().sum()
```

```python
# Correlation Matrix among all given features inside the dataframe

plt.figure(figsize=(30,20))
corr = data.corr()
sns.heatmap(corr, annot=True)
plt.savefig("Correlation Matrix.png")
plt.show()
```

```python
# Detection of outliers using Z-score

from scipy import stats
Z = np.abs(stats.zscore(data))
print(Z)
```

```python
# According to thumb rule if Z > 3 => This is an Outlier

print(np.where(Z>3))
```

```python
# Remove those reports whose Z-score > 3 
new_data = data[(Z < 3).all(axis=1)]
```

```python
new_data.shape
```

```python
# Split Data into Features(X) and target Columns(Y)

from sklearn.model_selection import train_test_split

# X - input feature
X = new_data.drop(columns='quality')

# Y - outcome of that feature
Y = new_data['quality']
```

```python
X.head()
```

```python
Y
```

```python
# Splitting

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
# Data  Modelling using Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
rf_classificationModel = RandomForestClassifier(n_estimators=100)
```

```python
rf_classificationModel.fit(X_train, Y_train)
```

```python
# Prediction of Data using Random Forest Model

y_pred = rf_classificationModel.predict(X_test)
```

```python
y_pred
```

```python
# Data Evaluation on Metrics

from sklearn import metrics
print('Accuracy Score',metrics.accuracy_score(Y_test,y_pred))
```

```python
from sklearn import tree

# Access the first decision tree in the forest
decision_tree = rf_classificationModel.estimators_[0]

# Plot the decision tree
plt.figure(figsize=(20, 10))
tree.plot_tree(decision_tree, filled=True)
plt.show()
```

```python

plt.figure(figsize=(20, 20))

# Iterate over all estimators (decision trees) in the Random Forest model
for i, decision_tree in enumerate(rf_classificationModel.estimators_):
    plt.figure(figsize=(20, 20))
    tree.plot_tree(decision_tree, filled=True)
    plt.title(f"Decision Tree {i + 1}")
    plt.show()
```
